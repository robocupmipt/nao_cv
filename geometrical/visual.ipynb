```python
import numpy as np
import cv2
import time
from IPython.display import clear_output
from pathlib import Path

%run math_calibration.ipynb

def put_text (img, text, str_num):
    font                   = cv2.FONT_ITALIC
    bottomLeftCornerOfText = (10, 50 + 30 * str_num)
    fontScale              = 0.8
    fontColor              = (15, 15, 215)
    lineType               = 2
    
    cv2.putText(img, text, 
        bottomLeftCornerOfText, 
        font, 
        fontScale,
        fontColor,
        lineType)

def output_calib_data (calib_data, frame, str_num):
    fr = frame.copy ()
    
    if (calib_data is not None and calib_data != 0):
        for key in calib_data.keys ():
            if (key == "opt_len" or
                key == "optical"):
                continue

            if (key == "corners"):
                for i in range (3):
                    x = calib_data [key] [i, 0]
                    y = calib_data [key] [i, 1]

                    cv2.circle (fr, (int (x), int (y)), 7, (123, 234, 145), -1)

                continue

            #print (key + ": ", calib_data [key])
            put_text (fr, key + ": " + str (np.around (calib_data [key], 1)), str_num)
            str_num += 1
    
    return fr

inp_path = "/Users/elijah/Dropbox/Programming/RoboCup/nao_cv/geometrical/chessboard_images/"

def main ():
    files = sorted (Path(inp_path).glob('**/*.jpg'))
    
    photo_num  = 0
    
    #print (photo_path)
    
    cv2.namedWindow ("frame")
    
    #cv2.resizeWindow ("frame", (480, 640))

    str_num    = 0
    calib_data = 0
    
    refresh_photo = True
    
    while (True):
        if (refresh_photo == True):
            photo_path = files [photo_num]
            frame = cv2.imread (str (photo_path))
            refresh_photo = False
        
        str_num = 0

        cv2.waitKey (1)    

        sz = 7
        shape = (sz, sz)
        
        calib_data_ = calibration_data (frame, shape, 38)

        if (calib_data_ is not None):
            calib_data = calib_data_ [0]

        frame = output_calib_data (calib_data, frame, str_num)

        frame__ = cv2.resize (frame, (480, 640))
        
        cv2.imshow ("frame", frame)

        time.sleep (0.02)

        clear_output (wait=True)
        
        keyb = cv2.waitKey (1) & 0xFF
        
        if (keyb == ord('q')):
            break

        if (keyb == ord('o')):
            photo_num -= 1
            refresh_photo = True

        if (keyb == ord('p')):
            photo_num += 1
            refresh_photo = True

    cv2.destroyAllWindows()

if __name__ == "__main__":
    main ()
```
